```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

req = Request('https://www.youtube.com/results?search_query=%EB%A9%9C%EB%A1%A0+2018%EB%85%84+11%EC%9B%94+4%EC%A3%BC%EC%B0%A8')
res = urlopen(req)
html = res.read().decode('utf-8')

bs = BeautifulSoup(html, 'html.parser')
tags = bs.findAll('div', attrs={'class': 'yt-lockup-content'})


for tag in tags:
    print(tag.find('a').attrs['href'] )
    print(tag.find('a').attrs['title'] )
    print('\n')
#    tag2 = tag.findAll('div', attrs={'class':'col_inner'} )
#    for tag3 in tag2:
#        
#        if (tag3.find('h4', attrs={'class':'mon'})!= None):
#            tag4 = tag3.findAll('div', attrs={'class':'thumb'})
#            for tag5 in tag4:
#                
#                print(tag5.find('a').attrs['href'])
#                print(tag5.find('img').attrs['src'])
#                print(tag5.find('img').attrs['title'])
#                print("\n")
```
