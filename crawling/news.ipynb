```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

req = Request('http://it.chosun.com/')
res = urlopen(req)
html = res.read().decode('utf-8')

bs = BeautifulSoup(html, 'html.parser')
tag = bs.find('div', attrs={'class':'thumb_news'})
tags = tag.findAll('li')

for tag in tags:
    b = tag.find('div', attrs={'class':'img'})
    print(b.find('a').attrs['href'])
    print(b.find('img').attrs['src'])
    b = tag.find('div', attrs={'class':'txt_wrap'})
    print(str(b.find('span', attrs={'class':'txt_dot1'}))[23:-7])
    b = tag.find('div', attrs={'class':'txt_wrap'})
    print(str(b.find('span', attrs={'class':'txt_dot2'}))[37:-7])
    print('\n\n\n')

#for tag in tags:
#    tag2 = tag.findAll('div', attrs={'class':'col_inner'} )
#    for tag3 in tag2:
#        
#        if (tag3.find('h4', attrs={'class':'mon'})!= None):
#            tag4 = tag3.findAll('div', attrs={'class':'thumb'})
#            for tag5 in tag4:
#                
#                print(tag5.find('a').attrs['href'])
#                print(tag5.find('img').attrs['src'])
#                print(tag5.find('img').attrs['title'])
#                print("\n")
```
